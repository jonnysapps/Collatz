# Observing The Collatz Conjecture

Below is a piece of code which prints the integer values between 1 and *n*:

```python
def Func(x):
    return x + 1

n = 5
x = 0
while x < n:
    x = Func(x)
    print(x)
```

Below two more numerical functions are defined:

```python
def Func2(x):
    return x//2

def Func3(x):
    return (3*x)+1
```

We will now write a similar piece of code to where we printed the integers between 1 and *n* but this time we are going to apply Func2/3 depending on whether *n* is even or odd respectively, stopping only if *n* is 1:

```python
n = 161
steps = 0
print(n)

while n != 1:
    if n % 2 == 0:
        n = Func2(n)
        print(n)
        steps = steps + 1
            
    else:
        n = Func3(n)
        print(n)
        steps = steps + 1
```

What we have above is a demonstration of one of the great unsolved mathematematical mysteries, the Collatz conjecture. It deems that as long as *n* is a positive integer the above code will always eventually get to 1 without entering any loops. No proof has ever been published so there my be a number which does not abide by these rules, none have yet been found.

The next thing we can look at is how the total number of iterations to reach 1 varies with *n*. This is done below:

```python
import pandas as pd

steps = 0
List = []
for n in range(1,200):
    num = n
    while n != 1:
        if n % 2 == 0:
            n = Func2(n)
            steps = steps + 1
            
        else:
            n = Func3(n)
            steps = steps + 1
    List.append([num,steps])
    steps = 0

print(List)

#plot results
%matplotlib inline

df = pd.DataFrame(List)
df.columns = ['n','Steps']

#print(df)

ax = df.plot(y = "Steps", x = "n")
ax.set_ylabel("Steps to reach 1")
```

It can be seen above that there is almost no patterns or trends to be drawn from the data.

Lets look at the top *steps* values and their corresponding *n* values to see if there is any relation between these numbers:

```python
df = df.sort_values('Steps', ascending = False)

print(df.head(10))
```

Once again this sheds very little light as the *n* values are a mixture of odd/even/prime/3-digit and 2-digit numbers. It can be noted that consecutive numbers can often yield the same result, but this could also been see from the above graph.

It could possibly be useful to see how *n* varies with step number for specific numbers of of interest. This can be found below: 

```python
import pandas as pd

n = 129
steps = 0
List2 = []
List2.append([steps,n])

while n != 1:
    if n % 2 == 0:
        n = Func2(n)
        steps = steps + 1
        List2.append([steps,n])
            
    else:
        n = Func3(n)
        steps = steps + 1
        List2.append([steps,n])

print(List2)

%matplotlib inline

df = pd.DataFrame(List2)
df.columns = ['Step','n']

ax = df.plot(y = "n", x = "Step")
ax.set_ylabel("n")
```
